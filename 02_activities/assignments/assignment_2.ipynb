```python
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:
    # TODO
    insert_pos = 0

    # Step 1: move non-zeros forward
    for num in nums:
        if num != 0:
            nums[insert_pos] = num
            insert_pos += 1

    # Step 2: fill remaining with zeros
    for i in range(insert_pos, len(nums)):
        nums[i] = 0

    return nums

```


-   Explain why their solution works in your own words.


```python
# This solution works by tracking the current position in the list (without 0s with insert_pos).
# Step 1 iterates through the list, checking each number to see if it is 0 or not. If the number is 0, the function continues
# without updating the "insert_pos" counter.
# If the number is anything else, the position of the current number is changed to [insert_pos] and the insert_pos counter is
# increased by 1. Effectively, this shifts each number over, removing all the 0s
# The second part adds the 0s back onto the end of the list by looping through the remaining positions in the list and changing 
# the value of each position to 0.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# The time complexity of this solution is O(2n). At a minimum, you will have to loop through the entire list once, and at a maximum
# you will have to loop through it twice (i.e. if all values are 0)

# The space complexity is 0(1) because nothing other than the counter is stored in memory
```
